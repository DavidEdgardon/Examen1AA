```
from collections import defaultdict

class Kruskal:
 
    def __init__(self, vertices): #C
        self.V = vertices
        self.graph = []    
     
    def addEdge(self, u, v, w): #C
        self.graph.append([u, v, w])
 
    def find(self, parent, i): 
        if parent[i] == i: 
            return i  
        return self.find(parent, parent[i]) 
 
    def union(self, parent, weight, x, y):
        xr = self.find(parent, x)  
        yr = self.find(parent, y)
 
        if weight[xr] < weight[yr]: 
            parent[xr] = yr 
        elif weight[xr] > weight[yr]:  
            parent[yr] = xr  
        else: 
            parent[yr] = xr  
            weight[xr] += 1 
 
    def MSTKruskal(self):
        result = []  
        i = 0 
        e = 0                
        self.graph =  sorted(self.graph, key=lambda item: item[2]) 
        parent, weight = [], [] 

        for node in range(self.V): 
            parent.append(node) 
            weight.append(0)
     
        while e < self.V-1:	
            u, v, w =  self.graph[i]
            i += 1
            x = self.find(parent, u)
            y = self.find(parent, v)
                                       
            if x != y:
                e += 1  
                result.append([u, v, w])
                self.union(parent, weight, x, y)          
 
        for u, v, weight in result: 
            print("%d <--> %d = %d" % (u, v, weight))
            
        total = 0
        for weight in result:   
            total += weight[2] 
        print("El árbol expandido mínimo con peso total de: ", total ) 
 
 
k = Kruskal(4)
k.addEdge(0, 1, 10)
k.addEdge(0, 2, 6)
k.addEdge(0, 3, 5)
k.addEdge(1, 3, 15)
k.addEdge(2, 3, 4)
k.addEdge(2, 4, 25)
 
k.MSTKruskal()
```

**Análisis de complejidad**

a es el número de aristas del grafo

n el número de vértices

El algoritmo de Kruskal requiere un tiempo que esta en **Θ(a log n)**. Para un grado denso, "a" tiende a n(n-1)/2, por lo que el algoritmo requiere un tiempo que esta Θ(n^2 log n). En un grafo disperso, "a" tiende a n-1, por lo que el algoritmo de Kruskal requiere un tiempo que esta en Θ(n log n)



**Analisis Cualitativo**

El algoritmo de Kruskal es un algoritmo que tiene como objetivo encontrar el minimo expancion del arbol con la mejor "ruta" posible entre los nodos. Basicamente lo que hace es, crea un conjunto S que contenga todas las aristas del arbol, quita un borde con un peso minimo de S, si el borde eliminado conecta 2 arboles diferentes,lo agrega al "bosque" F, combinando 2 arbol en un solo arbol. Al finalizar el algoritmo, el bosque forma un bosque de expansion minimo del grafico, si el grafico esta conectado, el "bosque" tiene un solo componente y forma un arbol de expasion minimo.
